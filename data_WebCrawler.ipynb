```python
#写入CSV文件中
import csv

def write_to_csv(video_info):
    with open('bilibili_videos.csv', 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        # 写入标题行
        writer.writerow(['视频期数','视频标题','视频简介','视频链接','视频ID','视频分类','up主','up主头像链接','播放量','弹幕数','评论数','收藏数','投币数','分享数','点赞量','踩的数量','热评/推荐信息','图片链接','发布时间','视频时长'])
        # 写入数据行
        for row in video_info:
            writer.writerow(row)


write_to_csv(vedio_info_list)
```
